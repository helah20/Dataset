## 🧪 اختبار صلاحية البيانات للتحليل العاملي (KMO و Bartlett)

لحساب صدق البنية (Construct Validity)، نستخدم:

- **KMO (Kaiser-Meyer-Olkin)**:
  - يقيس مدى ملاءمة البيانات للتحليل العاملي.
  - إذا كانت القيمة:
    - < 0.6 = ضعيف
    - 0.6 - 0.7 = متوسط
    - 0.7 - 0.8 = جيد
    - 0.8 - 0.9 = جيد جدًا
    - > 0.9 = ممتاز

- **Bartlett's Test of Sphericity**:
  - يختبر ما إذا كانت المصفوفة الارتباطية تختلف عن مصفوفة الهوية.
  - إذا كانت القيمة الاحتمالية (p-value) أقل من 0.05 → البيانات مناسبة للتحليل العاملي.



```python
!pip install factor_analyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity # Make sure to import the functions

# حساب KMO
kmo_all, kmo_model = calculate_kmo(df_num)

# حساب مقياس KMO لتحديد مدى ملاءمة البيانات للتحليل العاملي
# KMO > 0.6 تعتبر مقبولة، وكلما اقتربت من 1 كان أفضل
chi_square_value, p_value = calculate_bartlett_sphericity(df_num)

print(f"KMO: {kmo_model:.3f}")
print(f"Bartlett's Test Chi-Square: {chi_square_value:.3f}, p-value: {p_value:.5f}")
```

```python
def plot_bartlett(chi_square, p_value):
    fig, ax = plt.subplots(figsize=(5, 2))
    ax.axis('off')
    text = f"Bartlett’s Test\nChi-Square = {chi_square:.2f}\nP-Value = {p_value:.5f}"
    color = 'green' if p_value < 0.05 else 'red'
    plt.text(0.5, 0.5, text, fontsize=14, ha='center', va='center', bbox=dict(boxstyle="round", facecolor=color, alpha=0.3))
    plt.show()

plot_bartlett(chi_square_value, p_value)
```

#Normality


```python
import pandas as pd
import scipy.stats as stats

#
data = df_num

# Perform Shapiro-Wilk Test for normality
statistic, p_value = stats.shapiro(data)
print(f'Shapiro-Wilk Test statistic: {statistic:.3f}, p-value: {p_value:.3f}')

# Check if the data is normally distributed
if p_value > 0.05:
    print("The data is normally distributed.")
else:
    print("The data is not normally distributed.")
```

```python
import pandas as pd
import scipy.stats as stats

# Create an empty list to store the results
normality_results = []

# Loop through columns in the DataFrame
for column in df_num.columns:
    # Perform Shapiro-Wilk test for each column
    statistic, p_value = stats.shapiro(df_num[column])

    # Determine normality based on p-value
    is_normal = p_value > 0.05

    # Append the results to the list as a dictionary
    normality_results.append({'Column': column, 'Statistic': statistic, 'P-Value': p_value, 'Is Normal': is_normal})

# Create a DataFrame from the results list
normality_df = pd.DataFrame(normality_results)

# Display the DataFrame
print(normality_df)
```

```python
import scipy.stats as stats

# Loop through columns in the DataFrame
for column in data.columns:
    # Perform Kolmogorov-Smirnov test for each column
    statistic, p_value = stats.kstest(data[column], 'norm')
    print(f'Kolmogorov-Smirnov Test for column {column}:')
    print(f'  Statistic: {statistic:.3f}, p-value: {p_value:.3f}')

    # Check if the data is normally distributed for each column
    if p_value > 0.05:
        print("  The data in this column is normally distributed.")
    else:
        print("  The data in this column is not normally distributed.")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
import scipy.stats as stats

# Regression: Total Score ~ Password Awareness Score + Browser Score + Social Media Score
X = sm.add_constant(df[['Password Awareness Score', 'Browser Score', 'Social Media Score']])  # Independent variables + constant
y = df['Total Score']  # Dependent variable
model = sm.OLS(y, X).fit()

# Extract R-squared and Adjusted R-squared
r_squared = model.rsquared
adj_r_squared = model.rsquared_adj

# Pearson correlation coefficients
r_password, p_password = stats.pearsonr(df['Password Awareness Score'], df['Total Score'])
r_browser, p_browser = stats.pearsonr(df['Browser Score'], df['Total Score'])
r_social, p_social = stats.pearsonr(df['Social Media Score'], df['Total Score'])

# Breusch-Pagan test for heteroskedasticity
bp_test = het_breuschpagan(model.resid, X)
labels = ['LM Statistic', 'LM-Test p-value', 'F-Statistic', 'F-Test p-value']

# Print results
print("Regression Results:")
print(f"R-squared: {r_squared:.4f}")
print(f"Adjusted R-squared: {adj_r_squared:.4f}")
print("\nPearson Correlation Coefficients:")
print(f"r (Password Awareness Score vs Total Score): {r_password:.4f}, p-value: {p_password:.4f}")
print(f"r (Browser Score vs Total Score): {r_browser:.4f}, p-value: {p_browser:.4f}")
print(f"r (Social Media Score vs Total Score): {r_social:.4f}, p-value: {p_social:.4f}")
print("\nBreusch-Pagan Test Results:")
for label, value in zip(labels, bp_test):
    print(f"{label}: {value}")
```
